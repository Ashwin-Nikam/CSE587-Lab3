```r
require(gdata)
Frame = read.xls ("NHLTop100.xlsx", sheet = 1, header = TRUE)
```

Reading xlsx data

```r
Frame
```

```r
model <- lm(Frame$A ~ Frame$G, data = Frame)
model
summary(model)
```

```r
library(ggplot2)
gp <- ggplot(Frame, aes(x = G, y =  A))
gp <- gp + geom_point(aes(colour = GP, size = G)) + scale_colour_gradient(low = "red")
gp <- gp + geom_text(data=subset(Frame, G == 893 & A == 1962),
            aes(G,A,label=Player), vjust=-1) + xlim(0,950) + ylim(0,2100) + xlab("Goals") + ylab("Assists")
gp <- gp + scale_size_continuous(range = c(0, 4))
gp <- gp +  geom_abline(intercept = model$coefficients[1], slope = model$coefficients[2])
gp
```

Linear model of Goals vs Assists plotted on the graph where intercept:  494.5135  and slope:  0.5202

```r
plot(Frame$G, Frame$A)
abline(lm(I(Frame$A-1963) ~ I(Frame$G-894) +0, data=Frame), col="red")
model1 <-  lm(I(Frame$A-1963) ~ I(Frame$G-894) +0, data=Frame)
coef(model1)
abline(-437.39, 2.68, col="blue") #y intercept calculated using formula y = mx + b
model1
```

```r
plot(Frame1$G, Frame1$A)
model3 <- lm(Frame1$A ~ Frame1$G + 0, data = Frame1)
abline(lm(Frame$A ~ Frame$G + 0, data = Frame))
summary(model3)
model3
```

1. lm(Frame1$A ~ Frame1$G + 0, data = Frame1) gives us a regression line which passes through the origin as shown in the above figure
2. Now we want to plot a line which passes through origin and Wayne Gretzky
3. So here, we consider intercept = 0 and calculate slope by considering the equation y = mx + b where (x,y) are the coordinates of Wayne Gretzky. So we get the equation m = y/x as b = 0.
4. We similarly do this for Patrick Kane
5. We get the following two slopes 2.19 and 1.62

```r
gp <- ggplot(Frame1, aes(x = G, y =  A))
gp <- gp + geom_point(aes(colour = GP, size = G)) + scale_colour_gradient(low = "red")
gp <- gp + geom_text(data=subset(Frame1, G == 894 & A == 1963),
            aes(G,A,label=Player), vjust=-1) + xlim(0,950) + ylim(0,2100) + xlab("Goals") + ylab("Assists")
gp <- gp + geom_text(data=subset(Frame1, G == 285 & A == 462),
            aes(G,A,label=Player), vjust=-1)
gp <- gp + scale_size_continuous(range = c(0, 4))
gp <- gp + geom_abline(intercept = 0, slope = 2.19, color = "blue")
gp <- gp + geom_abline(intercept = 0, slope = 1.62, color = "red")
gp <- gp + geom_abline(intercept = 0, slope = model3$coefficients[1], color = "black")
gp
```
