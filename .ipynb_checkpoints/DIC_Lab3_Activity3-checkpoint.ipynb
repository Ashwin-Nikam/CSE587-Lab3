```r
Data <- read.csv(file="June 10-July 12, 2015 - Gaming, Jobs and Broadband - CSV.csv", header=TRUE, sep=",")
DFrame = as.data.frame(Data)
```

```r
names(DFrame)
```

```r
Frame <- cbind(DFrame$game1, DFrame$emplnw, DFrame$age, DFrame$eminuse,
         DFrame$sex)
Frame <- data.frame(Frame)
names(Frame) <- c("Gamer","Employed","Age","Use internet","Sex")
Frame
```

Filtered out 5 columns which are to be required for kmeans

```r
str(Frame)
```

```r
summary(Frame)
```

The summary of the frame shows that the data is varying in range by looking at the 'min' and 'max' of each value

```r
normalize <- function(x, na.rm = TRUE){
    ranx <- range(x, na.rm = na.rm)
    (x - ranx[1]) / diff(ranx)
}
```

Normalization function

```r
Frame <- as.data.frame(lapply(Frame, normalize))
```

```r
str(Frame)
```

```r
summary(Frame)
```

If we look at the data we can now see that it has been normalized as the 'min' and 'max' values are 0 and 1 respectively and all other values range between them 

```r
km <- kmeans(Frame, 3)
```

Kmeans carried out on Frame with 3 clusters

```r
km
```

```r
plot(x = Frame$Age, y = Frame$Employed, col = km$cluster)
points(km$centers[,c("Age", "Employed")], col=1:3, pch=23, cex=3)
```

```r
km5 <- kmeans(Frame, 5)
```

Now we use kmeans() with number of clusters to be created as 5

```r
km5
```

```r
plot(x = Frame$Age, y = Frame$Employed, col = km5$cluster)
points(km5$centers[,c("Age", "Employed")], col=1:3, pch=23, cex=3)
```

```r
rng<- c(1:21) #K from 2 to 20
tries <-100 #Run the K Means algorithm 100 times
avg.totw.ss <-integer(length(rng)) #Set up an empty vector to hold all of points
avg.bet.ss <- integer(length(rng))
for(v in rng){ # For each value of the range variable
 v.totw.ss <-integer(tries) #Set up an empty vector to hold the 100 tries
 v.bet.ss <-integer(tries) 
 for(i in 1:tries){
 k.temp <-kmeans(Frame,centers=v) #Run kmeans
 v.totw.ss[i] <-k.temp$tot.withinss#Store the total withinss
 v.bet.ss[i] <- k.temp$betweenss
 }
 avg.totw.ss[v-1] <-mean(v.totw.ss) #Average the 100 total withinss
 avg.bet.ss[v-1] <- mean(v.bet.ss)
}
plot(rng,avg.totw.ss,type="b", main="Between SS/ Total SS",
 ylab="Between SS and Total SS",
 xlab="Value of K", xlim =c(1,21),ylim = c(0,610))
lines(rng,avg.bet.ss,type="b", col="blue")
```

1. From the above graph we can see that the TotalSS decreases as we increase the size of the clusters.
2. The Between SS value increases as we increases the size
3. Thus the BetweenSS/TotalSS ratio will improve with increasing value of K
4. Here we have shown values of K ranging from 1 to 20
